```python
# print table schemas for all tables in mimic.db
import sqlite3
from pprint import pprint

# Connect to the mimic.db SQLite database
conn = sqlite3.connect('mimic_db/sample_mimic.db')
cursor = conn.cursor()

# Fetch all table names
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = [row[0] for row in cursor.fetchall()]

# Build a dictionary of table schemas
schema_dict = {}
for table in tables:
    cursor.execute(f"PRAGMA table_info({table});")
    columns = cursor.fetchall()
    schema_dict[table] = [
        {
            'cid': col[0],
            'name': col[1],
            'type': col[2],
            'notnull': bool(col[3]),
            'default_value': col[4],
            'pk': bool(col[5])
        }
        for col in columns
    ]

conn.close()

# Pretty-print the result
pprint(schema_dict)

```
