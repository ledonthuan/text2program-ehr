```python
# check sample_mimic.db to make sure the data is populated
import sqlite3
import pandas as pd

# Path to your SQLite file
db_path = 'mimic_sparql/mimicsql/evaluation/mimic_db/mimic.db'

# Connect to the database
conn = sqlite3.connect(db_path)
cursor = conn.cursor()

# Get list of all tables
cursor.execute("SELECT name FROM sqlite_master WHERE type='table';")
tables = [row[0] for row in cursor.fetchall()]
print(f"Found {len(tables)} tables:\n", tables)

# Check each table
empty_tables = []
for table in tables:
    try:
        # Count rows
        cursor.execute(f"SELECT COUNT(*) FROM {table};")
        row_count = cursor.fetchone()[0]
        print(f"Table: {table} - Rows: {row_count}")
        
        if row_count == 0:
            empty_tables.append(table)
        else:
            # Optional: Show first 5 rows
            df = pd.read_sql_query(f"SELECT * FROM {table} LIMIT 5;", conn)
            print(df.head())
            print('-' * 60)
    except Exception as e:
        print(f"Error reading table {table}: {e}")
        continue

# Summary
if empty_tables:
    print("\n⚠️ Empty tables found:", empty_tables)
else:
    print("\n✅ All tables contain data.")

# Clean up
cursor.close()
conn.close()
```
